## Demographics

```python
# age
my_bar(data, 'age', 'How old are you?', None, ['18 - 27', '28 - 37', '38 - 47', '48 - 57', '58 - 67'])
```

```python
fig, (ax1, ax2) = mpl.pyplot.subplots(ncols=2, figsize=(10, 5))
my_bar(data[data.variant == 'ai'], 'age', 'Age across AI variant', ax1, ['18 - 27', '28 - 37', '38 - 47', '48 - 57', '58 - 67'])
my_bar(data[data.variant == 'no-ai'], 'age', 'Age across no-AI variant', ax2, ['18 - 27', '28 - 37', '38 - 47', '48 - 57', '58 - 67'])
```

```python
# education level
my_bar(data, 'education', 'What is the highest education level you have completed?', None, 
    ['Secondary education', 'High school diploma', 'Technical/community college', 'Undergraduate degree', 'Graduate degree', 'Doctorate degree']
)
```
